# DATA PREPROCESSING AND DATA VISUALIZATION

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
Data=pd.read_csv('/home/s5ad1/DATA SCIENCE LAB/train_cars(1).csv')
```

```python
Data.head(5)
```

```python
Data.dtypes
```

```python
Data.isna().sum()
```

```python
Data['demand']=Data['demand'].fillna(Data['demand'].mean())
```

```python
Data.isna().sum()
```

```python
Data['date']=pd.to_datetime(Data['date'])
```

```python
Data['year']=Data['date'].apply(lambda x:x.year)
```

```python
Data
```

```python
Data['month']=Data['date'].apply(lambda x:x.month)
```

```python
Data
```

```python
Data['dayofweek']=Data['date'].apply(lambda x:x.dayofweek)
```

```python
Data
```

```python
Data['day']=Data['date'].apply(lambda x:x.day)
```

```python
Data['isquaterdate']=Data['date'].apply(lambda x:x.quarter)
```

```python
Data
```

```python
Data['isweekend']=Data['dayofweek'].apply(lambda x:1 if x>4 else 0)
```

```python
Data
```

```python
Data.drop('date',axis=1,inplace=True)
```

```python
Data
```

```python
def timeofday(n):
    if n in range(1,4):
        return 'Late night'
    if n in range(4,7):
        return 'Early morning'
    if n in range(7,12):
        return 'Morning'
    if n in range(12,3):
        return 'After noon'
    if n in range(3,18):
        return 'Evening'
    if n in range(18,25) or n==0:
        return 'Night'
    
```

```python
Data['timeofday']=Data['hour'].apply(lambda x:timeofday(x))
```

```python
Data
```

```python
def seasons(n):
    if n in [12,1,2]:
        return 'Winter'
    if n in [3,4,5]:
        return 'Summer'
    if n in [6,7,8]:
        return 'Monsoon'
    if n in [9,10,11]:
        return 'Autumn'
```

```python
Data['season']=Data['month'].apply(lambda x:seasons(x))
```

```python
sns.barplot(x=Data['season'],y=Data['demand'],palette='cubehelix')
```
